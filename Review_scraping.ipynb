```python
# Mengimpor pustaka google_play_scraper untuk mengakses ulasan dan informasi aplikasi dari Google Play Store.
from google_play_scraper import app, reviews, Sort
import pandas as pd
import csv
```

```python
# Mengambil semua ulasan dari aplikasi ChatGPT di Google Play
scrapreview, _ = reviews(
    'com.openai.chatgpt',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=50000
)
```

```python
# Convert ke DataFrame
df_reviews = pd.DataFrame(scrapreview)

print("Baris dan Kolom pada DataFrame:", df_reviews.shape)

df_reviews.head()
```

```python
# Simpen ke CSV
df_reviews.to_csv('ulasan_chatgpt.csv', index=False, encoding='utf-8')
```
